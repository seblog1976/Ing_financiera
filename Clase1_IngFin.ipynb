```python
def cf_series(face,cupon,maturity):
    cf=[]
    for c in range(maturity):
        cf.append(face*cupon)
    cf[-1]=cf[-1]+face
    return cf
        
    
```

```python
bono=cf_series(100,0.03,5)
```

```python
bono
```

```python
bono_2=cf_series(100,0.2,7)
```
